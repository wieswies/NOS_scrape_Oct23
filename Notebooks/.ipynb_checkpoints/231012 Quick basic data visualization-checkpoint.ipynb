## Dataset information
### NOS articles from October 4, 2022 up to October 4, 2023
Including the following metadata:
    Article ID
    Link
    Title 
    Date
    Time
    Category
    Images
    Paragraphs

```python
article_df = pd.read_excel('NOS_articles_221004-231004_corrected.xlsx')
```

```python
article_df = article_df.drop('Unnamed: 0', axis=1)
```

```python
print(article_df.shape)
article_df.dtypes
```

### Basic data manipulation

##### Add complementary date values

```python
article_df['Date'] = article_df['Date'].apply(pd.Timestamp)
article_df['Date'] = pd.to_datetime(article_df['Date'], format='%Y-%m-%d')
article_df['Month'] = article_df['Date'].dt.to_period('M')
article_df['Weekday'] = article_df['Date'].dt.weekday
```

##### Convert data types

```python
article_df['Date'] = pd.to_datetime(article_df['Date']).dt.date
article_df['Time'] = pd.to_datetime(article_df['Time'], format = '%H:%M:%S').dt.time
```

##### Extract main category

```python
nos_main_categories = [
    'Voorpagina',
    'Laatste nieuws',
    'Binnenland',
    'Buitenland',
    'Regionaal nieuws',
    'Politiek',
    'Economie',
    'Koningshuis',
    'Tech',
    'Cultuur & Media',
    'Opmerkelijk'
]
```

```python
# Remove characters
article_df['Category'] = article_df['Category'].str.replace('\\', '', regex=True).str.replace("'", '', regex=True)

# Extract main category
article_df['Main category'] = article_df['Category'].apply(lambda x: list(main for main in nos_main_categories if main in x), None).apply(lambda x: ', '.join(map(str, x)))
```

```python
# Check
article_df['Main category']
```

## Exploratory basic data visualization

```python
article_counts_date = article_df.groupby('Date')['Link'].count().reset_index()
```

```python
article_counts_date
```

#### Timeline

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=article_counts_date['Date'],
    y=article_counts_date['Link'],
    mode='lines+markers',
    hovertemplate='%{x|%A, %B %d, %Y}<br>Articles: %{y}<extra></extra>',
    line=dict(shape='linear'),
    marker=dict(size=8),
))

fig.update_layout(
    title='NOS articles; Oct 4, 2022 - Oct 4, 2023',
    xaxis_title='Date',
    yaxis_title='Number of Articles',
    xaxis=dict(
        tickformat="%b %Y",  
        tickangle=45,  
    )
)

fig.show()
```

#### Per news category

```python
article_df_catgraph = article_df.copy()
article_df_catgraph['str(Month)'] = article_df_catgraph['Month'].dt.strftime('%b %Y')
```

```python
grouped_data = article_df_catgraph.groupby(['Date', 'Main category']).size().unstack().fillna(0)
```

```python
cmap = plt.get_cmap('tab20', article_df_catgraph['Main category'].nunique())

plt.figure(figsize=(10, 6))

for i, category in enumerate(grouped_data.columns):
    plt.fill_between(grouped_data.index, 0, grouped_data[category], label=category, color=cmap(i), alpha = 0.4)

plt.xlabel('Month')
plt.ylabel('#Articles')
plt.title('Line Chart of Articles by Main Category')
plt.legend()
plt.grid(True)

plt.show()
```

#### Summarized category overview

```python
category_counts = article_df_catgraph.groupby('Main category')['Link'].count()

category_counts.sort_values(ascending=False)[:15]
```

### Focus on politics

```python
article_df_pol = article_df[article_df['Category'].str.contains('Politiek', case=False, na=False)]
```
